```python
kgcovid = pd.read_csv("./embedded_graph.embedded",index_col=0, sep=" ")
#kgcovid = pd.io.parsers.read_csv("embedded_graph.embedded",sep="\t",index_col=0)
kgcovid.head()
```

```python
import umap
import umap.plot
```

```python
reducer = umap.UMAP()
```

```python
embedding = reducer.fit_transform(kgcovid)
embedding.shape
```

```python
plt.scatter(
    embedding[:, 0],
    embedding[:, 1])
    #c=[sns.color_palette()[x] for x in penguins.species_short.map({"Adelie":0, "Chinstrap":1, "Gentoo":2})])
    
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of first KG-COVID embeddings', fontsize=24)
```
